# 1999-2015年美国各州吸毒死亡率

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False
death = pd.read_csv('drug_induced_deaths_1999-2015.csv')
death.dropna(subset = ['Crude Rate'], inplace = True)

plt.style.use('bmh')

# 绘制直方图
plt.hist(death['Crude Rate'],
        bins = np.arange(death['Crude Rate'].min(),
                         death['Crude Rate'].max(),
                         1),
        normed = True,
        color = 'r',
        alpha = 0.4,
        edgecolor = 'r', # 黑色
        )

plt.title('1999-2015年美国各州吸毒死亡率')
plt.xlabel('死亡率')
plt.ylabel('频率（%）')

# 绘制密度图
kde = mlab.GaussianKDE(death['Crude Rate'])
x2 = np.linspace(death['Crude Rate'].min(),
                death['Crude Rate'].max(),
                100)
line2 = plt.plot(x2, kde(x2), 'green', linestyle='--', linewidth = 2)
# 去除图形顶部边界和右边界的刻度
plt.tick_params(top='off', right='off')

plt.show()
```
