# 散点图矩阵

## （1）对角线上的图表用KDE方法绘制

```python
%matplotlib inline
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # matplotlib的拓展包

crime = pd.read_csv('crimeRatesByState2005.csv')

# 去掉异常点
crime2 = crime[crime.state != 'United States']
crime2 = crime2[crime.state != 'District of Columbia']
# 去除第一列（axis=0：删除行，axis=1：删除列）
crime2 = crime2.drop({"state"}, axis = 1)

## 绘制散点图矩阵
# 对角线上是各个属性的分布图，非对角线上是两个不同属性之间的相关图
q = sns.pairplot(crime2, # 数据
                 kind = 'scatter', # 散点图
                 diag_kind = 'kde', # 设置对角线图为密度图
                 plot_kws = dict(s=60, edgecolor='w', color='g', alpha=0.5), # 设置点的样式
                 diag_kws = dict(shade=True, color='y') # 设置对角线图的样式
                )

plt.show()
```

## （2）每个子图表都使用线性回归方法(REG)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # matplotlib的拓展包

crime = pd.read_csv('crimeRatesByState2005.csv')

# 去掉异常点
crime2 = crime[crime.state != 'United States']
crime2 = crime2[crime.state != 'District of Columbia']
# 去除第一列（axis=0：删除行，axis=1：删除列）
crime2 = crime2.drop({"state"}, axis = 1)

# 绘制散点图矩阵
# 对角线上的图表用KDE（Kernal Density Estimation）方法绘制
# 每个子图表都使用线性回归方法regression(REG)
sns.set(style="ticks", color_codes=True)
q = sns.pairplot(crime2, 
                 kind='reg', # 回归分布图
                 diag_kind = 'hist', # 设置对角线图为直方图
                 height=2, # 图表大小
                 markers='o',
                 plot_kws = dict(color='y'), # 设置点的样式
                 diag_kws = dict(fill=False, edgecolor='g') # 设置对角线图的样式
                )

plt.show()
```
