# 散点图

## 1. 使用ggplot进行绘制

### （1）基本绘制方法

```python
import ggplot as gp
import pandas as pd
import numpy as np

crime = pd.read_csv('crimeRatesByState2005.csv')
print(gp.ggplot(gp.aes(x='murder', y='burglary'), data=crime) # 设置x轴和y轴以及数据来源
     + gp.geom_point(color='red') # 利用 + 进行拼接，并设置颜色为红色
     )
```

### （2）去掉异常点

```python
import ggplot as gp
import pandas as pd
import numpy as np

crime = pd.read_csv('crimeRateByState2005.csv')
# 去除全美平均值和华盛顿特区这两个数据点
crime2 = crime[crime.state != 'United States']
crime2 = crime2[crime.state != 'District of Columbia']

print(gp.ggplot(gp.aes(x='murder', y='burglary', data=crime2))
     + gp.geom_point()
     + gp.stat_smooth(method='loess', color='red')
     )
```

## 2. 使用matplotlib进行绘制

### （1）基本绘制方法

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

# 读取数据
crime = pd.read_csv('crimeRatesByState2005.csv')

# 设置x轴和y轴的数据（将dataframe格式的数据转换为列表格式）
data_x = list(crime['murder'])
data_y = list(crime['burglary'])

# 创建一个图像
plt.figure()
# 绘制散点图
plt.scatter(data_x, data_y, s=50, c='seagreen', marker='o', alpha=0.6)

# 添加参考线
plt.grid(ls='--', c='black')

# 设置x，y轴标签
# 设置字体
font = {"family": "SimHei"}  
plt.xlabel('谋杀案', font)
plt.ylabel('入室盗窃案', font)
plt.title("散点图示例", font)

plt.show()
```

### （2）去除异常点

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

crime = pd.read_csv('crimeRatesByState2005.csv')
# 去除全美平均值和华盛顿特区两个数据点
crime2 = crime[crime.state != 'United States']
crime2 = crime2[crime.state != 'District of Columbia']

# 设置x轴和y轴的数据（将dataframe格式的数据转换为列表格式）
x = list(crime2['murder'])
y = list(crime2['burglary'])

# 创建一个图像
plt.figure()
# 绘制散点图
plt.scatter(x, y, s=60, c='deepskyblue', marker='o', alpha=0.6)

# 一元函数
poly = np.polyfit(x, y, deg=1)
py = np.polyval(poly, x)

# 原函数线沿y轴上下移动200形成的线
py1 = py + 200
py2 = py -200

# 绘制线
plt.plot(x, py, c='lightcoral', alpha=0.7)
plt.plot(x, py1, alpha=0) # 将线设置为透明的
plt.plot(x, py2, alpha=0) # 将线设置为透明的

# 将线py1和py2围成的区间填充颜色
plt.fill_between(x, py1, py2, where=py1>py2, facecolor='violet', alpha=0.2)

# 添加参考线
plt.grid(ls='--', c='grey')

# 设置x, y轴标签
font = {"family" : "KaiTi"}
plt.xlabel('谋杀案', font)
plt.ylabel('入室盗窃案', font)
plt.title("散点图示例(去除异常点)", font)

plt.show()
```
