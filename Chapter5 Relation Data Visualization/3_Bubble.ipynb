# 气泡图

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 导入数据
crime = pd.read_csv('crimeRatesByState2005.csv')

# 输出“谋杀”列的数值
print(list(crime.murder))

# 去除异常点
crime2 = crime[crime.state != 'United States']
crime2 = crime2[crime.state != 'District of Columbia']

# 指定图形的字体
font = {'family' : 'KaiTi',
        'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 16,
        }

# 选取一个z值
z = list(crime2.population / 10000)
# 产生服从“0~1”均匀分布的随机样本值
colors = np.random.rand(len(list(crime2.murder)))
# 设置scatter颜色渐变
cm = plt.cm.get_cmap('rainbow')

# 绘制散点图
im = plt.scatter(list(crime2.murder),
                list(crime2.burglary),
                s = z,
                c = z,
                cmap = cm,
                linewidths = 0.5,
                alpha = 0.5
               )

# 显示colorbar
cbar = plt.colorbar(im)
cbar.set_label('$colorbar$',fontdict=font)

# 设置x，y轴标签
plt.xlabel('murder')
plt.ylabel('burglary')
# 设置标题
plt.title("murder-burglary气泡图", font)

plt.show()
```
