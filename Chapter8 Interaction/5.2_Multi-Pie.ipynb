# 组合饼图

```python
from pyecharts.charts import Pie
from pyecharts.charts import Grid
import pyecharts.options as opts
import pandas as pd
from pyecharts.globals import ThemeType

# 定义根据城市获取其'AQI_Bucket'列数据的函数
# 需要传入三个参数：城市、圆心、标题位置
def get_pie(city, center1, center2):
    data = df.loc[df['City'] == city, ['Date', 'AQI_Bucket']]
    
    # 按'AQI_Bucket'进行分组
    rank_message = data.groupby(['AQI_Bucket'])
    # 对不同空气质量等级进行计数
    rank_com = rank_message['AQI_Bucket'].agg(['count'])
    # 重置索引
    rank_com.reset_index(inplace=True)
    # 按计数'count'降序排列
    rank_com_last = rank_com.sort_values('count', ascending=False)
    
    x = rank_com_last['AQI_Bucket']
    y = rank_com_last['count']
    
    # 设置画布大小
    pie = Pie(init_opts=opts.InitOpts(width='350px',
                                      height='300px',
                                      theme=ThemeType.MACARONS,
                                     )
             )
    input_data = [list(z) for z in zip(x, y)]
    
    # 添加数据，并设置饼图圆心位于图表的位置，以及内外圆的半径
    pie.add(city, input_data, center=center1, radius=['15%', '30%'])
    
    # 设置标签格式
    pie.set_series_opts(label_opts=opts.LabelOpts(formatter='{b}: {d}%'),
                       )
    
    # 设置标题和图例格式
    pie.set_global_opts(title_opts=opts.TitleOpts(title=city,
                                                  # 设置饼图标题位于图表的位置
                                                  pos_left=center2[0],
                                                  pos_top=center2[1],
                                                 ),
                        legend_opts=opts.LegendOpts(is_show=False)
                       )
    return pie

df = pd.read_csv('city_day.csv')
cities = ['Ahmedabad', 'Chennai', 'Delhi', 'Lucknow']
# 标题的位置（以指定位置为起点）
pos1 = [['25%', '25%'], ['25%', '75%'], ['75%', '25%'], ['75%', '75%']]
# 饼图的位置（以指定位置为中点）
pos2 = [['19%', '23%'], ['22%', '74%'], ['73%', '23%'], ['71%', '74%']]

# 设置grid大小、主题
grid = Grid(init_opts=opts.InitOpts(width='1020px',
                                    height='800px',
                                    theme=ThemeType.CHALK,
                                   ),
           )

# 将饼图依次添加到grid中
for i in range(len(cities)):
    grid.add(get_pie(cities[i], pos1[i], pos2[i]),
             grid_opts=opts.GridOpts(),           
            )
    
grid.render_notebook()
```
