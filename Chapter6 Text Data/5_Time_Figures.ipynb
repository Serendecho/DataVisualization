# 《时代周刊》年度人物来自的领域的词云图

```python
from pyecharts.charts import WordCloud   # 导入画词云图需要的库
import pyecharts.options as opts
import pandas as pd

# 读取数据文件
post_data = pd.read_csv('Person of the Year.csv')

# 通过链式调用对数据进行处理：
post_data2 = post_data.groupby(by=['Category']).agg({'Count': sum}).reset_index()

# 通过for循环将数据转化为二维列表
data = [list(e) for e in zip(post_data2['Category'], 
                             [str(e) for e in post_data2['Count']]
                            )
       ]

# 绘制词云图
wordcloud = WordCloud()
wordcloud.add(series_name = '', # 系列名称
              data_pair = data, # 词和词频组成的数据对，二维列表的格式
              word_size_range = [20, 100], # 设置词云字体大小的区间
              shape = 'star', # 词云图的轮廓
              word_gap = 18, # 单词间隔
              rotate_step = 45, # 旋转单词角度
              textstyle_opts=opts.TextStyleOpts(font_family="times new roman", font_size=14),
             )

wordcloud.render_notebook()
```
