# 词云图

```python
from pyecharts.charts import WordCloud   # 导入画词云图需要的库
import pyecharts.options as opts
import pandas as pd

# 读取数据文件
post_data = pd.read_csv('post_data.csv')

## 通过链式调用对数据进行处理：
# grouby：进行分组操作，by=['category']表示按种类进行分组；
# agg：aggregate，对列进行操作的函数，agg({'views': sum})是将每组中的浏览量相加求和
# reset_index()：重置索引（因为按种类分组后之前的索引就不能用了）
post_data2 = post_data.groupby(by=['category']).agg({'views': sum}).reset_index()

# 通过for循环将数据转化为二维列表
data = [list(e) for e in zip(post_data2['category'], [str(e) for e in post_data2['views']])]

# 绘制词云图
wordcloud = WordCloud()
wordcloud.add(series_name = '', # 系列名称
              data_pair = data, # 词和词频组成的数据对，二维列表的格式
              word_size_range = [20, 100], # 设置词云字体大小的区间
              shape = 'diamond', # 词云图的轮廓
              word_gap = 15, # 单词间隔
              rotate_step = 45, # 旋转单词角度
             )

wordcloud.render_notebook()
```
