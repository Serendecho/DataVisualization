# 主题河流图

### type_的取值
（1）value：数值轴，适用于连续数据

（2）category：类目轴，适用于离散的类目数据，为该类型时，必须通过data设置类目数据

（3）time：时间轴，适用于连续的时序数据，与数值轴相比，时间轴带有时间的格式化，在刻度计算上也有所不同

```python
from pyecharts.charts import ThemeRiver
import pyecharts.options as opts
import csv

x_data = [] # x轴数据列表
y_data = [] # y轴数据列表

# 读取数据文件
with open('data.csv', 'r') as f:
    r_csv = list(csv.reader(f))   #
    for row in r_csv[1:]:  # 
        # 添加y轴的数据，列表的形式
        y_data.append([row[0], int(row[1]), row[2]]) # 将num列的字符串转化为int型
        
        # 获得不重复的主题标签，并添加到x轴数据列表中
        if row[2] not in x_data:
            x_data.append(row[2])
            
## 绘制主题河流图
(
    # 初始化时设置图像的宽度和高度
    ThemeRiver(init_opts=opts.InitOpts(width='1000px', height='700px'))
    
    .add(
        series_name = x_data,  # 输入标签名列表
        data = y_data,         # 输入数据列表
        # 
        singleaxis_opts = opts.SingleAxisOpts(
            pos_top = '50',
            pos_bottom = '50',
            type_ = 'time', # 设置类型为时间轴
        ),
    )
    
    # 在全局设置中添加坐标轴提示框（当鼠标放到图像上时会出现一条参考线）
    .set_global_opts(
        tooltip_opts = opts.TooltipOpts(trigger='axis', axis_pointer_type='line')
    )
    
    # 渲染图像
    .render_notebook()
)
```
