# 不同病的死亡率

```python
from pyecharts.charts import ThemeRiver
import pyecharts.options as opts
import csv
from pyecharts.globals import ThemeType

x_data = [] # x轴数据列表
y_data = [] # y轴数据列表

# 读取数据文件
with open('death_rate.csv', 'r') as f:
    r_csv = list(csv.reader(f)) 
    for row in r_csv[1:]:  # 
        # 添加y轴的数据，列表的形式
        # 将Age Adjusted Death Rate列的字符串转化为float型
        y_data.append([row[1], float(row[2]), row[0]]) 
        
        # 获得不重复的主题标签，并添加到x轴数据列表中
        if row[0] not in x_data:
            x_data.append(row[0])
            
## 绘制主题河流图
(
    # 初始化时设置图像的宽度和高度
    ThemeRiver(init_opts=opts.InitOpts(width='880px', height='780px', theme=ThemeType.VINTAGE))
    
    .add(
        series_name = x_data,  # 输入标签名列表
        data = y_data,         # 输入数据列表
        # 设置单轴配置项
        singleaxis_opts = opts.SingleAxisOpts(
            pos_top = '55',
            pos_bottom = '55',
            type_ = 'time', # 设置类型为时间轴
        ),
    )
    
    # 在全局设置中添加坐标轴提示框（当鼠标放到图像上时会出现一条参考线）
    .set_global_opts(
        tooltip_opts = opts.TooltipOpts(trigger='axis', axis_pointer_type='line'),
        legend_opts=opts.LegendOpts(pos_top='10')
    )
    
    # 渲染图像
    .render_notebook()
)
```
