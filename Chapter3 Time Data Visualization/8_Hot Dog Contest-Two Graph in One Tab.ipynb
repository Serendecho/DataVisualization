# 将两个图表放在一起——使用Tab

```python
from pyecharts.charts import Tab
from pyecharts.charts import Bar
from pyecharts.globals import ThemeType
import pyecharts.options as opt
import csv

## 创建一个Tab实例
tab = Tab()

## 创建图表1
bar1 = Bar({"theme": ThemeType.WONDERLAND})
filename1 = './chapter3-time data/hot-dog-contest-winners.csv'
data1x = []
data1y = []
with open(filename1) as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            data1x.append(datarow[0])
            data1y.append(float(datarow[2]))
bar1.add_xaxis(data1x)
bar1.add_yaxis('大胃王冠军的成绩', data1y, label_opts=opt.LabelOpts(is_show=False))
bar1.set_global_opts(title_opts=opt.TitleOpts(title='历年大胃王冠军的成绩'),
                    xaxis_opts=opt.AxisOpts(name="年份"),
                    yaxis_opts=opt.AxisOpts(name="大胃王冠军的成绩"),
                    brush_opts=opt.BrushOpts())
bar1.set_series_opts(markpoint_opts=opt.MarkPointOpts(
                    data=[
                        opt.MarkPointItem(type_="max", name="最大值"),
                        opt.MarkPointItem(type_="min", name="最小值"),
                        opt.MarkPointItem(type_="average", name="平均值"),
                    ]
                    ))

## 创建图表2
bar2 = Bar({"theme": ThemeType.CHALK})
filename2 = './chapter3-time data/hot-dog-places.csv'
data2x = []
data2y = []
with open(filename2) as f:
    reader = csv.reader(f)
    for datarow in reader:
        data2x.append(datarow)
x2 = data2x[0]
y21 = [float(x) for x in data2x[1]]
y22 = [float(x) for x in data2x[2]]
y23 = [float(x) for x in data2x[3]]

bar2.add_xaxis(x2)
bar2.add_yaxis('A', y21, stack = 1)
bar2.add_yaxis('B', y22, stack = 2)
bar2.add_yaxis('C', y23, stack = 1)
bar2.set_series_opts(label_opts=opt.LabelOpts(is_show=False))

## 将两个图表放入Tab中
tab.add(bar1, 'Bar')   # 标签为'Bar'
tab.add(bar2, 'Stack') # 标签为'Stack'

tab.render_notebook()
```
