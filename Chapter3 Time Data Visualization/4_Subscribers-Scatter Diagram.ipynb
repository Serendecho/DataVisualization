# 微信公众号订阅量的散点图

```python
%matplotlib inline
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib

# matplotlib默认显示英文，要显示中文需要进行设置
matplotlib.rc("font", family='KaiTi', weight="bold")

# 读取csv文件
filename = './chapter3-time data/flowingdata_subscribers.csv'
datay = []
with open(filename) as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            datay.append(int(datarow[1]))

# 获取横坐标的值
xa = list(range(1, len(datay)+1))
# 画图：设置
plt.scatter(xa, datay, s = 30, c = 'g', marker = 'p', alpha = 0.6, label = "订阅量")
plt.legend(loc = "lower right")   # 添加图例，位置在右下方
plt.title("微信公众号的订阅量")   # 添加标题
plt.show()                        # 展示图像
```
