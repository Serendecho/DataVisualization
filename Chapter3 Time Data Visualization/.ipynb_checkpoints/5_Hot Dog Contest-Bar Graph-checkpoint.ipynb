# 获胜者比赛成绩的柱形图

## pyecharts的主题样式设置
from pyecharts.globals import ThemeType

bar = Bar({"theme": ThemeType.XXX})

其中，XXX可以换为：

 |  CHALK = 'chalk' #粉笔风
 |  
 |  DARK = 'dark'  #暗黑风
 |  
 |  ESSOS = 'essos' #厄索斯大陆
 |  
 |  INFOGRAPHIC = 'infographic' #信息图
 |  
 |  LIGHT = 'light' #明亮风格
 |  
 |  MACARONS = 'macarons' #马卡龙
 |  
 |  PURPLE_PASSION = 'purple-passion' #紫色激情
 |  
 |  ROMA = 'roma' #石榴
 |  
 |  ROMANTIC = 'romantic' #浪漫风
 |  
 |  SHINE = 'shine' #闪耀风
 |  
 |  VINTAGE = 'vintage' #复古风
 |  
 |  WALDEN = 'walden' #瓦尔登湖
 |  
 |  WESTEROS = 'westeros' #维斯特洛大陆
 |  
 |  WHITE = 'white' #洁白风
 |  
 |  WONDERLAND = 'wonderland' #仙境

```python
from pyecharts.charts import Bar
import pyecharts.options as opt
import csv
from pyecharts.globals import ThemeType

bar = Bar({"theme": ThemeType.VINTAGE})  # 创建一个柱形图的实例

# 打开csv文件
filename = './chapter3-time data/hot-dog-contest-winners.csv'
datax = []
datay = []
with open(filename) as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            datay.append(float(datarow[2])) # 数据集中读取的数据为字符串，需要转换为float型
            datax.append(datarow[0])        # pyecharts中横坐标要求为字符串，故不进行转换

# 将从文件中读取的数据添加到柱形图的参数中
bar.add_xaxis(datax)
bar.add_yaxis('大胃王冠军的成绩', datay, label_opts=opt.LabelOpts(is_show=True)) # 柱形图上显示具体数据
bar.set_global_opts(title_opts=opt.TitleOpts(title='历年大胃王冠军的成绩'),  # 设置柱形图的标题
                    xaxis_opts=opt.AxisOpts(name="年份"),
                    yaxis_opts=opt.AxisOpts(name="大胃王冠军的成绩"),
                    datazoom_opts=opt.DataZoomOpts(),                        # 提供区域缩放配置项
                    brush_opts=opt.BrushOpts())                              # 提供区域选择组件

bar.render_notebook()
```
