```python
from pyecharts.charts import Line, Bar, Grid
import pyecharts.options as opts
import csv
from pyecharts.globals import ThemeType

## 打开csv文件
filename = 'scores-smart product.csv'
datax = []
datay1 = []
datay3 = []
datay4 = []

# 读入数据
try:
    with open(filename) as f:
        reader = csv.reader(f)
        for datarow in reader:
            if reader.line_num != 1:
                datax.append(datarow[0])
                datay1.append(int(datarow[1]))   # Recs
                datay3.append(int(datarow[3]))   # TLCS
                datay4.append(int(datarow[4]))   # TGCS            
except:
    print("Fail to open the file!")
            
## 生成折线图
line = Line({"theme": ThemeType.ROMA})   # 设置主题样式
# 将从文件中读取到的数据添加到x轴中
line.add_xaxis(datax)
# 将TGCS的数据添加到y0轴中
line.add_yaxis("TGCS", datay4, is_symbol_show = True, 
               symbol = "circle",yaxis_index=0,
               markpoint_opts=opts.MarkPointOpts( # 添加标记点
                   data=[opts.MarkPointItem(type_="min", name="最小值"),
                         opts.MarkPointItem(type_="max", name="最大值"),
                         opts.MarkPointItem(type_="average", name="平均值")
                        ]))
# 将TLCS的数据添加到y1轴中
line.add_yaxis("TLCS", datay3, is_symbol_show = True, 
               symbol = "rectangle",yaxis_index=1,
               markpoint_opts=opts.MarkPointOpts( # 添加标记点
                   data=[opts.MarkPointItem(type_="min", name="最小值"),
                         opts.MarkPointItem(type_="max", name="最大值"),
                         opts.MarkPointItem(type_="average", name="平均值")
                        ]))
# 将Recs的数据添加到y2轴中
line.add_yaxis("Recs", datay1, is_symbol_show = True, 
               symbol = "triangle", yaxis_index=2,
               markpoint_opts=opts.MarkPointOpts( # 添加标记点
                   data=[opts.MarkPointItem(type_="min", name="最小值"),
                         opts.MarkPointItem(type_="max", name="最大值"),
                         opts.MarkPointItem(type_="average", name="平均值")
                        ]))

# 在右侧添加y轴
line.extend_axis(yaxis=opts.AxisOpts(name="TLCS",
                                    min_=0,
                                    max_=520,
                                    position="right"))
line.extend_axis(yaxis=opts.AxisOpts(name="Recs",
                                    min_=0,
                                    max_=520,
                                    offset=61, # y轴相对于默认位置的偏移
                                    position="right"))
# 设置全局参数
line.set_global_opts(title_opts=opts.TitleOpts(title="Recs & TLCS & TGCS"), # 设置标题
                    yaxis_opts=opts.AxisOpts(name="TGCS",
                                             position="left",
                                             min_=0,
                                             max_=5200,
                                            ),
                     # 添加交叉指向工具
                    tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross",
                                                 is_show=True),
                    )

line.render_notebook()
```
