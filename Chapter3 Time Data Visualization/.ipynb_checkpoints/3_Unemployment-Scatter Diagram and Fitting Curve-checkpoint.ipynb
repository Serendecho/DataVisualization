# 失业率的散点图及拟合曲线

matplotlib.rc("font",family='KaiTi',weight="bold")

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import csv
import sys  # 用于控制程序

# matplotlib默认显示英文，要显示中文需要自己进行设置
matplotlib.rc("font",family='KaiTi',weight="bold")

filename = './chapter3-time data/unemployment-rate-1948-2010.csv'
ya = []
xa = []
try:
    with open(filename) as f:
        reader = csv.reader(f)
        for datarow in reader:
            if reader.line_num != 1:
                ya.append(float(datarow[3]))
                xa.append(int(datarow[1]))
                
except:
    print('Error reading csv file')
    sys.exit(-1)

# 创建一个图像并设置其大小  
plt.figure(figsize = (10, 7), dpi = 80)  

# 在图像上画一个散点图（点的大小为15，颜色为红色，点形状为o，透明度为0.45）
plt.scatter(xa[:], ya[:], s=25, c='r', marker='o', alpha=0.45, label="失业率（unemployment rate）")
plt.legend(loc='lower right')  # 标签位置在右下角

# 添加x轴、y轴、标题信息
plt.xlabel("年份（Year）")
plt.ylabel("失业率（Unemployment Rate）")
plt.title("1948-2010年失业率")

# 在图像上画一条拟合曲线
poly = np.polyfit(xa, ya, deg=3)   # 用于求拟合曲线，设置为三次函数 
plt.plot(xa, np.polyval(poly, xa)) # 根据拟合函数求出对应的拟合曲线
plt.show()
```
