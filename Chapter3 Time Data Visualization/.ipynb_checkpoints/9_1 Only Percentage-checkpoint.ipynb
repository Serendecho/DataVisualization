```python
from pyecharts.charts import Line
import pyecharts.options as opts
import csv
# from pyecharts.globals import ThemeType

# 打开csv文件
filename = 'scores-smart product.csv'
datax = []
datay2 = []

try:
    with open(filename) as f:
        reader = csv.reader(f)
        for datarow in reader:
            if reader.line_num != 1:
                datax.append(datarow[0])
                datay2.append(float(datarow[2]))   # Percentage
            
except:
    print("Fail to open the file!")
            
# 生成折线图
line = Line()
# 将从文件中读取到的数据添加到参数中
line.add_xaxis(datax)
line.add_yaxis("Percentage", datay2, is_smooth = True, is_symbol_show = True, 
               symbol = "circle")
line.set_global_opts(title_opts=opts.TitleOpts(title='Cited Percentage in Area of Smart Watch',
                                              pos_top='5%', pos_left='center'),
                    yaxis_opts=opts.AxisOpts(axisline_opts=opts.AxisLineOpts(
                                                linestyle_opts=opts.LineStyleOpts(width=2, color='green')),
                                            axistick_opts=opts.AxisTickOpts(is_show=True,length=15),
                                            splitline_opts=opts.SplitLineOpts(is_show=True,
                                                                             linestyle_opts=opts.LineStyleOpts(color='green'))
                                            )
                    )

line.render_notebook()
```
