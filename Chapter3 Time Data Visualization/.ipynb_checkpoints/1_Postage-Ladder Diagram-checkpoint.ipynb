# 美国邮费阶梯图

```python
from pyecharts.charts import Line
import pyecharts.options as opts
from pyecharts.globals import ThemeType

line = Line({"theme": ThemeType.WESTEROS})                 # pyecharts 将阶梯图视为特殊的折线图
datax = ['1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009']   # 横坐标必须是字符串
datay = [0.32, 0.32, 0.32, 0.32, 0.33, 0.33, 0.34, 0.37, 0.37, 0.37, 0.37, 0.39, 0.41, 0.42, 0.44]
line.add_xaxis(datax)
line.add_yaxis('Price', datay, is_step = True) # is_step 为真时，画的是阶梯图，否则是折线图  

line.set_global_opts(title_opts = opts.TitleOpts(title = '美国邮费阶梯图'),
                     yaxis_opts = opts.AxisOpts(name = "Price", min_ = 0.3, max_ = 0.45),
                     xaxis_opts = opts.AxisOpts(name = "Year"),
                     datazoom_opts=opts.DataZoomOpts(),                        # 提供区域缩放配置项
                     brush_opts=opts.BrushOpts()
                    ) 

line.render_notebook()  # 如果是render就跳转网页，如果是notebook就在本网页显示
```
