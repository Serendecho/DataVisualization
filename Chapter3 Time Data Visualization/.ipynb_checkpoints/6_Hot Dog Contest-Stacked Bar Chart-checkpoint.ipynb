# 前三名成绩的堆叠柱形图

```python
from pyecharts.charts import Bar
import pyecharts.options as opt
from pyecharts.globals import ThemeType
import csv

# 创建一个柱形图的实例
bar = Bar({"theme": ThemeType.ROMA})

# 打开并读取csv文件
filename = './chapter3-time data/hot-dog-places.csv'
datax = []
datay = []
with open(filename) as f:
    reader = csv.reader(f)
    for datarow in reader:
        datax.append(datarow)
            
# 将读取的数据作为参数
x = datax[0]
y1 = [float(x) for x in datax[1]]
y2 = [float(x) for x in datax[2]]
y3 = [float(x) for x in datax[3]]

bar.add_xaxis(x)
bar.add_yaxis('A', y1, stack = 1) # stack = 0时，不会堆叠
bar.add_yaxis('B', y2, stack = 2) # stack数值相同时（除了0），会堆叠在一起
bar.add_yaxis('C', y3, stack = 1)
bar.set_global_opts(title_opts=opt.TitleOpts(title='历年大胃王比赛前三名的成绩'),
                   toolbox_opts=opt.ToolboxOpts())   # 添加工具箱
bar.set_series_opts(label_opts=opt.LabelOpts(is_show=False))

bar.render_notebook()
```
