# 将两个图表放在一起——使用page()作为容器

```python
from pyecharts.charts import Bar
from pyecharts.charts import Page
from pyecharts.globals import ThemeType
import pyecharts.options as opt
import csv

page = Page()  # 创建一个Page实例作为容器

## 生成柱形图1
bar1 = Bar({"theme": ThemeType.WESTEROS})    # 创建一个Bar实例
# 读取csv文件
filename1 = './chapter3-time data/hot-dog-contest-winners.csv'
data1x = []
data1y = []
with open(filename1) as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            data1y.append(float(datarow[2]))
            data1x.append(datarow[0])
bar1.add_xaxis(data1x)
bar1.add_yaxis('大胃王冠军成绩', data1y, label_opts=opt.LabelOpts(is_show=True))
bar1.reversal_axis()
bar1.set_series_opts(label_opts=opt.LabelOpts(position = "right"))
bar1.set_global_opts(title_opts=opt.TitleOpts(title='历年大胃王冠军成绩'))

## 生成柱形图2
bar2 = Bar({"theme": ThemeType.ROMA})
filename2 = './chapter3-time data/hot-dog-places.csv'
data2x = []
data2y = []
with open(filename2) as f:
    reader = csv.reader(f)
    for datarow in reader:
        data2x.append(datarow)
x2 = data2x[0]
y21 = [float(x) for x in data2x[1]]  
y22 = [float(x) for x in data2x[2]]
y23 = [float(x) for x in data2x[3]]

bar2.add_xaxis(x2)
# y轴数据堆叠
bar2.add_yaxis('A', y21, stack = 1)
bar2.add_yaxis('B', y22, stack = 1)
bar2.add_yaxis('C', y23, stack = 1, is_selected = False)

bar2.set_series_opts(label_opts=opt.LabelOpts(is_show=False),
                    markline_opts=opt.MarkLineOpts(
                        data=[
                            opt.MarkLineItem(type_="min", name="最小值"),
                            opt.MarkLineItem(type_="max", name="最大值"),
                            opt.MarkLineItem(type_="average", name="平均值"),
                        ]))

# 将两个图放在一个page里
page.add(bar1)
page.add(bar2)

page.render_notebook()
```
