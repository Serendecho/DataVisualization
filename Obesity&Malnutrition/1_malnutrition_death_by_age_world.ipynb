```python
from pyecharts.charts import Timeline, Pie
import pyecharts.options as opts
import csv
from pyecharts.globals import ThemeType

# 定义绘制轮播饼图的函数
def timeline_pie(attr, year, datay):
    attr = attr
    tl = Timeline(init_opts=opts.InitOpts(width='1050px',height='600px',theme=ThemeType.CHALK))
    tl.add_schema(is_auto_play = True,
                  play_interval = 600, # 设置播放的速度
                  is_loop_play = True,  # 循环播放
                  width=750,
                  pos_left='center',
                  pos_bottom='-1px',
                 )
    for i in range(len(year)):
        y = datay[i]
        input_data = [list(z) for z in zip(attr, y)]  # 数据对
        pie = Pie(init_opts=opts.InitOpts(width='600px',height='450px',theme=ThemeType.CHALK))
        pie.add("",
                input_data,
               )
        pie.set_global_opts(title_opts=opts.TitleOpts(title="{}年全球各年龄段因营养不良死亡的人数".format(year[i]),
                                                      pos_left='center',
                                                      pos_top='5',
                                                      title_textstyle_opts=opts.TextStyleOpts(font_size=26),
                                                     ),
                            legend_opts=opts.LegendOpts(orient='vertical',
                                                        type_="scroll",
                                                        pos_top='12%',
                                                        pos_left='80%',
                                                       ),
                           )
        pie.set_series_opts(label_opts=opts.LabelOpts(formatter='{b}: {d}%',
                                                      font_size=14,
                                                     ),
                           )
        tl.add(pie, "{}年".format(year[i]))
    return tl

attr = ['Under 5', '5-14 years', '15-49 years', '50-69 years', '70+ years']
year = []  # 获取年份数据
datay = [] # 获取各年龄段因营养不良死亡的人数数据

# 读取文件
with open('malnutrition-deaths-by-age_world.csv', 'r') as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            # 获取年份数据
            year.append(datarow[0])
            # 获取各年龄段因营养不良死亡的人数数据
            datay.append(datarow[1:])
            
# 绘制轮播饼图
timeline_pie(attr, year, datay).render_notebook()
```
