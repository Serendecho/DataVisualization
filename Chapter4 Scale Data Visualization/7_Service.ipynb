```python
from pyecharts.charts import Pie
import pyecharts.options as opts
import pandas as pd
from pyecharts.globals import ThemeType

# 读取文件
df = pd.read_csv('2019_service_industry.csv')
# 转换数据格式
data = [list(z) for z in zip(df['行业'], df['收入利润'])]

# 创建Pie的实例并设置画布的大小
pie = Pie(init_opts=opts.InitOpts(width="850px", height="600px", theme=ThemeType.MACARONS))  
# 添加数据
pie.add('', 
        data, 
        center=['45%', '60%'], 
        radius=['30%', '75%'],  
       )

# 设置全局元素
pie.set_global_opts(title_opts=opts.TitleOpts(title='2019年中国服务业各行业收入利润',
                                             # 标题位置
                                             pos_top="20",  
                                             pos_left='center',
                                             # 设置字体大小
                                             title_textstyle_opts=opts.TextStyleOpts(font_size='22'), 
                                             ),
                   legend_opts=opts.LegendOpts(pos_left='86%',  # 图例位置在右边
                                               pos_top='12%',
                                               orient='vertical',  # 垂直放置
                                              ),
                   )

# 设置标签的格式
pie.set_series_opts(label_opts=opts.LabelOpts(formatter='{b}: {d}%',
                                             font_size='14',   # 设置标签的字体大小
                                             ),
                   )

pie.render_notebook()
```
