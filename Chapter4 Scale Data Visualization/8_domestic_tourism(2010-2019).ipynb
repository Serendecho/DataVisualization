# 2010-2019年中国旅游业情况

```python
from pyecharts.charts import Bar
import pyecharts.options as opts
from pyecharts.globals import ThemeType
import pandas as pd

# 读取文件
tourism = pd.read_csv('2010-2019_domestic_tourism.csv')

# 添加一个Bar实例，并添加3个x轴
bar = Bar(init_opts=opts.InitOpts(width='950px', height='550px', theme=ThemeType.ESSOS))
bar.add_xaxis(['城乡旅游人数','城乡旅游总花费','城乡人均花费'])

# 添加y轴数据
# 在第一列的所有行中进行循环遍历
for i in range(tourism.iloc[:, 0].size): 
    # issue：行标签索引为i且列标签索引为‘年份’的数据
    issue = tourism.loc[i, '年份']   
    
    # y轴标签为issue
    # 数据为行标签索引为i且列标签索引分别为：
    # '城乡旅游人数','城乡旅游总花费','城乡人均花费'的数据
    bar.add_yaxis(issue, [int(x) for x in tourism.loc[i, ['城乡旅游人数','城乡旅游总花费','城乡人均花费']]])
    
bar.set_global_opts(title_opts=opts.TitleOpts(title='2010-2019年中国国内旅游情况统计',
                                              subtitle='城乡旅游人数(千万人次)，'
                                              '城乡旅游总花费(百亿元)，城乡人均花费(元)',
                                              pos_left='center',    # 标题放在中间
                                              pos_top='10',
                                             ),
                   legend_opts=opts.LegendOpts(pos_left='90%',   # 图例放在右边
                                               pos_top='8%',
                                               orient='vertical',   # 且垂直放置
                                              ),
                   )
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False,       # 显示标签 
                                             ),
                   )

bar.render_notebook()
```
