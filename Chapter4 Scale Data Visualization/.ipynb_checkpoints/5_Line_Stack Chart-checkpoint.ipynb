# 堆叠折线图

## 1. 基本示例

```python
from pyecharts.charts import Line
import pyecharts.options as opts
import pandas as pd

year_population_age = pd.read_csv('./chapter4-scale data/us_population_by_age.csv')

x = [str(x) for x in year_population_age['year']]
y1 = [float(x) for x in year_population_age['year_under5']]
y2 = [float(x) for x in year_population_age['year5_19']]
y3 = [float(x) for x in year_population_age['year20_44']]
y4 = [float(x) for x in year_population_age['year45_64']]
y5 = [float(x) for x in year_population_age['year65above']]

line = Line()
line.add_xaxis(x)
line.add_yaxis('5岁以下', y1, is_smooth=True, stack=1)
line.add_yaxis('5至19岁', y2, is_smooth=True, stack=1)
line.add_yaxis('20至44岁', y3, is_smooth=True, stack=1)
line.add_yaxis('45至64岁', y4, is_smooth=True, stack=1)
line.add_yaxis('65岁以上', y5, is_smooth=True, stack=1)

line.set_global_opts(title_opts=opts.TitleOpts(title='人口老龄化'),
                    yaxis_opts=opts.AxisOpts(
                        splitline_opts=opts.SplitLineOpts(is_show=True), # 设置分割线
                                            ),
                    legend_opts=opts.LegendOpts(pos_left='right',
                                               orient='vertical',
                                               ),
                    )

line.set_series_opts(areastyle_opts=opts.AreaStyleOpts(opacity=0.5),  # 设置背景颜色的透明度 
                    label_opts=opts.LabelOpts(is_show=True),
                    )
line.render_notebook()
```

## 2. 调整画布的尺寸

```python
from pyecharts.charts import Line
import pyecharts.options as opts
import pandas as pd

year_population_age = pd.read_csv('./chapter4-scale data/us_population_by_age.csv')

x = [str(x) for x in year_population_age['year']]
y1 = [float(x) for x in year_population_age['year_under5']]
y2 = [float(x) for x in year_population_age['year5_19']]
y3 = [float(x) for x in year_population_age['year20_44']]
y4 = [float(x) for x in year_population_age['year45_64']]
y5 = [float(x) for x in year_population_age['year65above']]

# 设置画布尺寸
line = Line(init_opts=opts.InitOpts(width='880px',
                                   height='720px',
                                   ),
           )

# 添加数据
line.add_xaxis(x)
line.add_yaxis('5岁以下', y1, is_smooth=True, stack=1)
line.add_yaxis('5至19岁', y2, is_smooth=True, stack=1)
line.add_yaxis('20至44岁', y3, is_smooth=True, stack=1)
line.add_yaxis('45至64岁', y4, is_smooth=True, stack=1)
line.add_yaxis('65岁以上', y5, is_smooth=True, stack=1)

line.set_global_opts(title_opts=opts.TitleOpts(title='人口老龄化',
                                               pos_left='center',
                                               pos_top='2%'
                                              ),
                     # 设置分割线
                    yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True),
                                             axistick_opts=opts.AxisTickOpts(is_show=True),
                                             min_=0,
                                             max_=120,
                                            ),
                     # 设置图例格式
                    legend_opts=opts.LegendOpts(pos_left='90%',
                                                pos_top='8%',
                                                orient='vertical',
                                               ),
                     # 添加交叉指向工具
                     tooltip_opts=opts.TooltipOpts(trigger='axis', 
                                                   axis_pointer_type='cross', 
                                                   is_show=True,
                                                  )
                    )

line.set_series_opts(areastyle_opts=opts.AreaStyleOpts(opacity=0.5),  # 设置背景颜色的透明度 
                    label_opts=opts.LabelOpts(is_show=True),
                    )
line.render_notebook()
```
