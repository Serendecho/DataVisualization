# 1. 树图

```python
import json
from pyecharts.charts import Tree
import pyecharts.options as opts

# 读取文件
with open('./chapter4-scale data/GDP_data.json', 'r', encoding='utf-8') as f:
    j = json.load(f)
    data = [j]

# 创建一个Tree实例，设置绘图尺寸，以及背景色
tree = Tree(init_opts=opts.InitOpts(width='850px',
                                   height='700px',
                                   bg_color="#107a68",
                                   ),
            )

# 添加数据
tree.add('', data, collapse_interval=10)
tree.set_global_opts(title_opts=opts.TitleOpts(title='TreeMap-基本示例', 
                                               pos_left='center', 
                                               pos_top='25',
                                              )
                    )
tree.set_series_opts(label_opts=opts.LabelOpts(color="black"))
tree.render_notebook()
```

# 2. 矩形树图

## 2.1 基本示例 

```python
import json
from pyecharts.charts import TreeMap
import pyecharts.options as opts

with open('./chapter4-scale data/GDP_data_1.json', 'r', encoding='utf-8') as f:
    j = json.load(f)
    data = [j]

# 创建一个TreeMap实例，并设置画布大小
map = TreeMap(init_opts=opts.InitOpts(width='1000px',
                                     height='700px',
                                     ),
             )

map.add('GDP', 
        data, 
        label_opts=opts.LabelOpts(position='inside'), # 设置标签位置在矩形内部
       )

# 显示图例，并设置标题的样式
map.set_global_opts(legend_opts=opts.LegendOpts(is_show=True, pos_left="83%", pos_top="45"),
                    title_opts=opts.TitleOpts(title="世界人口GDP", pos_left="10%", pos_top="20"),
                   )

map.render_notebook()
```

## 2.2 初始只展示第一层

```python
import json
from pyecharts.charts import TreeMap
import pyecharts.options as opts

with open('./chapter4-scale data/GDP_data_1.json', 'r', encoding='utf-8') as f:
    j = json.load(f)
    data = [j]

# 创建一个TreeMap实例，并设置画布大小
map = TreeMap(init_opts=opts.InitOpts(width='850px', height='700px'))

map.add('GDP',
       data,
       label_opts=opts.LabelOpts(position='inside'),# 设置标签位置在矩形内部
       leaf_depth=1, # 设置初始的分支深度，这里的1表示初始只显示第一层
       drilldown_icon='▶▶', # 设置当节点可以下钻时的提示符
       
        # TreeMap-Levels-配置
       levels = [
           opts.TreeMapLevelsOpts(
               treemap_itemstyle_opts=opts.TreeMapItemStyleOpts(
                   border_color='#555',border_width=4, gap_width=4
               )
           ),
           opts.TreeMapLevelsOpts(
               color_saturation=[0.1, 0.4],
               treemap_itemstyle_opts=opts.TreeMapItemStyleOpts(
                   border_color_saturation=0.7, gap_width=2, border_width=2
               )
           ),
           opts.TreeMapLevelsOpts(
               color_saturation=[0.3, 0.7],
               treemap_itemstyle_opts=opts.TreeMapItemStyleOpts(
                   border_color_saturation=0.6, gap_width=1
               )
           ),
           opts.TreeMapLevelsOpts(color_saturation=[0.3, 0.5])
       ]
    )

# 不显示图例，并设置标题的样式
map.set_global_opts(legend_opts=opts.LegendOpts(is_show=False),
                    title_opts=opts.TitleOpts(title="世界人口GDP", pos_left="center", pos_top="20"),
                   )

map.render_notebook()
```
