# 18-24岁青少年自杀率

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 导入数据
suicide = pd.read_csv('suicide_18-24.csv')

# 指定图形的字体
font = {'family' : 'KaiTi',
        'color'  : 'black',
        'weight' : 'normal',
        'size'   : 16,
        }

# 选取一个z值
z = list(suicide.Year/ 5)
# 产生服从“0~1”均匀分布的随机样本值
colors = np.random.rand(len(list(suicide.Female)))
# 设置scatter颜色渐变
cm = plt.cm.get_cmap('terrain')

# 绘制散点图
im = plt.scatter(list(suicide.Female),
                list(suicide.Male),
                s = z,
                c = z,
                cmap = cm,
                linewidths = 1,
                alpha = 0.5
               )

# 显示colorbar
cbar = plt.colorbar(im)
cbar.set_label('$colorbar$',fontdict=font)

# 设置网格线
plt.grid(linestyle='-.')

# 设置x，y轴标签
plt.xlabel('Female')
plt.ylabel('Male')
# 设置标题
plt.title("18-24岁青少年自杀率（1999-2013年）", font)

plt.show()
```
